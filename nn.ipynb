```python
class value:
    def __init__ (self, data): 
        self.data = data
    def __repr__(self):
        return f"value(data={self.data})"
    def __add__(self, other):
        out = value(self.data + other.data)
        return out
    def __mul__(self, other):
        out = value(self.data * other.data)
        return out


a = value(2.0)
b = value(-3.0)
c = value(10.0)
a * b


```
